```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
client = MongoClient('localhost', 27017)
```

```python
db = client['vacancy_database']
```

```python
vacancies = db.vacancies
```

```python
df = pd.read_csv('hh_vacancies.csv')
df.head()
```

```python
df_dict = df.to_dict()
```

```python
#vacancies.delete_many({})
```

```python
vacancies.insert_many(df_dict)
```

```python
for vac in vacancies.find({}):
    pprint(vac)
```

#### Написать функцию, которая производит поиск и выводит на экран вакансии с заработной платой больше введённой суммы.

```python
min = float(input('Введите минимальную границу зарплаты: '))
```

```python
for vacancy in vacancies.find( {'$or': [{ 'Min_salary': { "$gt": min }}, {'Max_salary': { "$gt": min }}]}, {'Vacancy Name': 1, 'Min_salary': 1, '_id': 0} ):
    pprint(vacancy)
```

#### Написать функцию, которая будет добавлять в вашу базу данных только новые вакансии с сайта

```python
for raw in df_dict:
    vacancies.update_one({'Link': raw['Link']},
                        {'$set': {'Vacancy Name': raw['Vacancy Name'],
                                  'Link': raw['Link'],
                                  'Min_salary': raw['Min_salary'],
                                  'Max_salary': raw['Max_salary'],
                                  'Currancy': raw['Currancy'],
                                  'Source': raw['Source'],
                                 }}, upsert=True)
```
